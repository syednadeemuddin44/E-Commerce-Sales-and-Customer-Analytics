# E-Commerce Sales & Customer Analytics

### End-to-End Data Analytics Project

This project analyzes an e-commerce sales dataset to understand product performance, customer behavior, regional sales, profitability, discount impact, and sales trends using Python.

# 1. Project Overview

## Dataset Description

The dataset contains transactional records of an e-commerce business, including information about customers, products, sales, profit, discounts, shipping details, and geographical locations.

The objective of this project is to perform end-to-end data analysis and extract meaningful business insights that can support better business decisions.

## Business Problems

The analysis aims to answer the following business problems:

- Which product categories contribute the highest sales and profit?
- Which regions and states perform well or underperform?
- Which customer segments generate the highest revenue?
- How do discounts impact profitability?
- How do sales and profits change over time?

## Business Objectives

The project focuses on the following objectives:

1. Analyze product and sub-category performance.
2. Assess regional sales and profit performance.
3. Evaluate customer segment performance.
4. Analyze the impact of discounts on profitability.
5. Study sales and profit trends over time.
